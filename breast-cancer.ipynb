```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
df=pd.get_dummies(df, drop_first=True, dtype=int)
```

```python
df
```

```python
P=df.drop('diagnosis', axis=1)
q=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
P_train, P_test, q_train, q_test=train_test_split(P, q, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
P_train=sc.fit_transform(P_train)
P_test=sc.transform(P_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```python
model=Sequential()

model.add(Dense(units=128, activation='relu', input_dim=P_train.shape[1]))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(units=64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(units=32, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(P_train, q_train, batch_size=32, epochs=100, validation_data=(P_test, q_test))
```

```python
y_pred=model.predict(P_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, ConfusionMatrixDisplay


cm=confusion_matrix(q_test, y_pred.round())
print(cm)
print(f'Accuracy: {accuracy_score(q_test, y_pred.round())}')
print(f'Precision: {precision_score(q_test, y_pred.round())}')
print(f'Recall: {recall_score(q_test, y_pred.round())}')
print(f'F1-Score: {f1_score(q_test, y_pred.round())}')
cm_display=ConfusionMatrixDisplay(cm)
cm_display.plot()
plt.show()
```
